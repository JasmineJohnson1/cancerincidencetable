```python
def assignICDgroup(word):
    
    if word == 'C00-C97':
        icd_code = 'All sites'
    elif word == 'C43':
        icd_code = 'Melanoma of the skin'
    elif word == 'C44':
        icd_code = 'Skin, non-melanoma'
    elif word == 'C45':
        icd_code = 'Mesothelioma'
    elif word == 'C47':
        icd_code = 'Autonomic nervous system'
    elif word == 'C50':
        icd_code = 'Breast'
    elif word == 'C69':
        icd_code = 'Eye'
    elif word == 'C73':
        icd_code = 'Thyroid gland'
    elif word == 'C39' or word == 'C76' or word == 'C80': 
        icd_code = 'Other or unspecified',
    elif word >= 'C00' and word <= 'C14':
        icd_code = 'Mouth, pharynx'
    elif word >= 'C15' and word <= 'C26':
        icd_code = 'Digestive organs'
    elif word >= 'C00' and word <= '97':
        icd_code = 'All sites'
    elif word >= 'C60' and word <= 'C63':
        icd_code = 'Male genital organs'
    elif (word >= 'C30' and word <= 'C30') or word == 'C38':
        icd_code = 'Digestive organs'
    elif word == 'C40' or word == 'C41':
        icd_code = 'Digestive organs'
    elif word == 'C48' or word == 'C49':
        icd_code = 'Soft tissues'
    elif word >= 'C51' and word <= 'C58':
        icd_code = 'Female genital organs'
    elif word >= 'C64' and word <= 'C68':
        icd_code = 'Urinary organs'
    elif word >= 'C70' and word <= 'C72':
        icd_code = 'Central nervous system'
    elif word >= 'C81' and word <= 'C96':
        icd_code = 'Lymphoid/haematopoietic tissue'
    else:
        icd_code = 'not_found'
    return (icd_code)
```

```python
def formatICDcodes(ini_word):
    word = ini_word.split(".")[0]
    word_list = word.split("-")
    if len(word_list)>1:
        removal_word = word_list[1]
        word_list[1] = removal_word.replace(removal_word[0], '')  ##remove C
    
    for word in word_list:
        count = 0
        #word_list[count] = word.replace('.', '')
        if 'excl' in word:
            word_list[count] = word.replace('excl', '')
        count = count + 1
    if len(word_list) == 1:
        return word_list[0]
    else:
        return word_list[0] + '–' + word_list[1]
```

```python
def createDictionary(df):
    dic = {}
    for l in df['ICD-10']:
        dic[l] = (df.loc[df['ICD-10'] == l,'Cancer site'].values[0])
    return dic
```
